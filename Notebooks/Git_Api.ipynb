```python
import requests
```

```python
a = input()
```

```python
url = "https://api.github.com/users/"+a+"/repos"
response = requests.get(url)
response_dict = response.json()
print(response_dict)
```

```python
len(response_dict)
```

```python
repos = []
for i in range(len(response_dict)-1):
   repos.append(response_dict[i]["html_url"])
```

```python
repos
```

```python
repo_name = [x.replace('https://github.com/'+a+'/', '').replace('.jpg', '') for x in repos]
```

```python
repo_name
```

```python
url= []
for i in repo_name:
  url.append('https://api.github.com/repos/'+ a+ '/'+ i+'/languages')
```

```python
url
```

```python
languages = []
for i in range(len(url)):  
  response = requests.get(url[i])
  response_dict = response.json()
  languages.append(response_dict)
```

```python
for i in languages:
  s = sum(i.values())
  for k, v in response_dict.items():
    pct = v * 100.0 / s
    print(k, pct)
```
